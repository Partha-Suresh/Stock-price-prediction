The date entries are not in the same format in the .csv file. 
Changing that to dd-mm-yyyy format

```python

# Lists of years, months (names and numbers)
year = ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']
month = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
month_num = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

# Load your data
data = pd.read_csv('../../data/Stock Regression/TSLA Historical Data 2015-2024.csv')

# Initialize a dictionary to store the results
month_data = {}
filtered_data=[]
# Loop over the years and months
for i in range(0, len(year)):
    for j in range(0, len(month_num)):
        # Create the filter string (month and year)
        filter_date = month_num[j] + '-' + year[i]
        # print(filter_date)
        # filtered_data = data[data['Date'].str[3:10] == filter_date]
        dates=data['Date']
        
        change=data['Change']
        # print(dates.size)
        for k in range(0,dates.size):
            if(dates.loc[k][3:10] == filter_date):
                filtered_data.append(change[k])
        # Select the 'Change' column and store it in the dictionary
        # Storing a list of 'Change' values under the key month-year (e.g., '01-2024')
        month_data[filter_date] = filtered_data  # Convert the 'Change' column to a list

# Now you can access the 'Change' values for any month-year combination from the dictionary
print(month_data)

```
